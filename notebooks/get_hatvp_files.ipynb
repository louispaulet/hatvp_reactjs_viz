## Export list of declarants present in the dataset and their repsective xml file URLs  

This is the lightweight and fast option.  
It works as long as the free CORS proxy is running.  

```python
def make_xml_url(declaration_filename):
  try:
    return 'https://www.hatvp.fr/livraison/dossiers/' + declaration_filename
  except:
    # case of publications posted but not yet published
    return None

dataset_df['xml_url'] = dataset_df.apply(lambda row: make_xml_url(row['open_data']), axis=1)

# export dataset to csv
export_df = dataset_df[dataset_df['xml_url'].notna()][[
    'civilite',
    'prenom',
    'nom',
    'type_mandat',
    'qualite',
    'type_document',
    'departement',
    'date_publication',
    'date_depot',
    'xml_url'
    ]]

export_df['searchable_field'] = (
    export_df['civilite'] + ' ' + export_df['prenom'] + ' ' + export_df['nom'] + ' ' +
    export_df['type_mandat'] + ' ' + export_df['qualite'] + ' ' + export_df['type_document'] + ' ' + export_df['departement'] +
    export_df['date_publication'] + ' ' + export_df['date_depot']
)
export_df = export_df[['searchable_field', 'xml_url']]
export_df.to_csv('people_in_dataset_and_xml_urls.csv', index=False)
```
